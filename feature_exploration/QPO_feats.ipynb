```python
import numpy as np
import matplotlib.pyplot as plt
import os
import sys

from sklearn.manifold import TSNE
from sklearn.cluster import HDBSCAN
plt.rcParams['figure.figsize'] = [15, 10]
```

```python
qpos = np.loadtxt(os.getcwd()[:-19]+'feature_extraction/QPO_sim_lcsoft_dt100_feats.csv',delimiter=',')
qpoh = np.loadtxt(os.getcwd()[:-19]+'feature_extraction/QPO_sim_lchard_dt100_feats.csv',delimiter=',')
qpof = np.loadtxt(os.getcwd()[:-19]+'feature_extraction/QPO_sim_lcfull_dt100_feats.csv',delimiter=',')
npos = np.loadtxt(os.getcwd()[:-19]+'feature_extraction/nPO_sim_lcsoft_dt100_feats.csv',delimiter=',')
npoh = np.loadtxt(os.getcwd()[:-19]+'feature_extraction/nPO_sim_lchard_dt100_feats.csv',delimiter=',')
npof = np.loadtxt(os.getcwd()[:-19]+'feature_extraction/nPO_sim_lcfull_dt100_feats.csv',delimiter=',')
```

```python
qpo_feats = np.concatenate((qpof,qpos,qpoh),axis=1)
npo_feats = np.concatenate((npof,npos,npoh),axis=1)

all_feats = np.concatenate((qpo_feats,npo_feats))
```

```python
tsne_transform = TSNE().fit_transform(all_feats)
```

```python
plt.scatter(tsne_transform[:10000,0],tsne_transform[:10000,1],alpha=0.25,color='r',label='QPO')
plt.scatter(tsne_transform[10000:,0],tsne_transform[10000:,1],alpha=0.25,color='b',label='nPO')
plt.legend(loc='upper right')
plt.show()
```

```python
fig, axs = plt.subplots(1,2)
axs[0].scatter(tsne_transform[:10000,0],tsne_transform[:10000,1],alpha=0.25,color='r')
axs[1].scatter(tsne_transform[10000:,0],tsne_transform[10000:,1],alpha=0.25,color='b')
plt.show()
```

```python
hdb_clust = HDBSCAN(min_cluster_size=20)
hdb_clust.fit(all_feats)
print(max(hdb_clust.labels_))
```

```python
plt.scatter(tsne_transform[np.where(hdb_clust.labels_ == -1)[0],0],tsne_transform[np.where(hdb_clust.labels_ == -1)[0],1],alpha=0.25,color='k',label='Outliers')
for i in range(4):
    clust_idxs = np.where(hdb_clust.labels_ == i)
    plt.scatter(tsne_transform[clust_idxs,0],tsne_transform[clust_idxs,1],alpha=0.25,label='Clust '+str(i))
plt.legend(loc='upper right')
plt.show()
```
